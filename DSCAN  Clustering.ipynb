# DBSCAN --
Density Based CLustering of Application Noise

```python
import pandas as pd 
import numpy as np 
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
X =np.array([[1,2],[2,2],[2,3],[8,7],[8,8],[25,80]])
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
db =DBSCAN(eps=3,min_samples=2)
db.fit(X)
db.labels_
```

```python
from sklearn.datasets import make_circles
X,_ = make_circles(n_samples=500,factor=.6,noise=.03,random_state=4)

# Apply DBSCAN
dbscan =DBSCAN(eps=0.1,min_samples=5)
clusters =dbscan.fit_predict(X)

# Plotting
plt.scatter(X[:,0],X[:,1],c=clusters,cmap='viridis',marker='o')
plt.title('DSCAN ')
plt.xlabel('Feature1')
plt.ylabel('feature2')
plt.show()
```
