# Agglomerative Clustering 
a type of Hierarchical Clustring

```python
import pandas as pd 
import numpy as np 
```

```python
df =pd.read_csv('dataset\hierarchical-clustering-with-python-and-scikit-learn-shopping-data.csv')
```

```python

data =df.iloc[:,3:5]
data.head()
```

```python
import scipy.cluster.hierarchy as shc 
import matplotlib.pyplot as plt

plt.figure(figsize=(10,7))
plt.title('Customer Dendogram')
dend = shc.dendrogram(shc.linkage(data,method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster =AgglomerativeClustering(n_clusters=5,linkage='ward')
label_cluster =cluster.fit_predict(data)
```

```python
plt.figure(figsize=(10,7))
plt.scatter(data.iloc[:,0],data.iloc[:,1],c=label_cluster ,cmap='rainbow')
plt.show()
```
