# Grid SearchCV

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler ,OneHotEncoder
from sklearn.impute import SimpleImputer 
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
df =pd.read_csv('dataset\\train.csv')
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
X = df.drop(columns=['Survived'])
Y =df['Survived']
```

```python
X_train ,X_test,Y_train ,Y_test =train_test_split(X,Y,test_size=0.2,random_state=4)
```

```python
X_test.head()
```

```python
numerical_features =['Fare','Age',]
```

```python
numerical_transformer = Pipeline(memory=None ,steps=[('imputer',SimpleImputer(strategy='median'))
                                        ,('scacler',StandardScaler())],verbose=False)

```

```python
categorical_features =['Embarked','Sex']
```

```python
categorical_transformer =Pipeline(memory=None ,steps=[('imputer',SimpleImputer(strategy='most_frequent')),
                                         ('ohe',OneHotEncoder(handle_unknown='ignore'))],verbose=False)
```

```python
preproccer =ColumnTransformer(transformers=[('num',numerical_transformer,numerical_features),
                                            ('cat',categorical_transformer,categorical_features)],remainder='passthrough')
```

```python
#classifer = LogisticRegression()

clf =Pipeline(memory=None ,steps=[('preproccer',preproccer),('classifer',LogisticRegression())],verbose=False)
```

```python
from sklearn import set_config
set_config(display='diagram')
clf
```

# apply GridsearchCV

```python
param_grid ={'preproccer_num_imputer_straget':['mean','median'],
             'preproccer_cat_imputer_stragey':['most_frequent','constant'],
             'classifer':[0.1,1.0,10,100]}
grid_search =GridSearchCV(clf,param_grid,cv=10)
```

```python
grid_search.fit(X_train,Y_train)
print("Best Params :")
print(grid_search.best_params_)
```
