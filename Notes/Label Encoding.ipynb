Label Encoding - Converting the data into numeric
 form

```python
# importing the dependecies
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

Label Encoding of breast cancer data set

```python
cancer_data =pd.read_csv('/content/data.csv')
```

```python
cancer_data.head()
```

```python
cancer_data.shape
```

```python
#finding the count of different level
cancer_data['diagnosis'].value_counts()
```

```python
#load the LabelEncoder
label_encode =LabelEncoder()
```

```python
labels = label_encode.fit_transform(cancer_data.diagnosis)
```

```python
print(labels)
```

```python
cancer_data['target'] = labels
```

```python
cancer_data.head()
```

0 ---> Benign

0----->Maligant

```python
cancer_data['target'].value_counts()
```
