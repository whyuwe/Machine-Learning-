```python
# importing the library
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Note : Seaborn has some built-in dataset

```python
# total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.shape
```

```python
# setting a theme for plots
sns.set_theme()
```

```python
# Visualize the data
sns.relplot(data = tips, x ='total_bill',y = 'tip',col ='time',hue ='smoker',style ='smoker',size ='size')
```

```python
from sklearn.datasets import load_breast_cancer
digit = load_breast_cancer()
df = pd.DataFrame(digit.data ,columns = digit.feature_names)
df['Price'] = digit.target
df.head()
```

HeatMap

```python
#  Constructing Heat Map
plt.figure(figsize =  (10,10))   # (figure = (size))
sns.heatmap(df , cbar = True ,square = True, fmt = '.1f', annot = True ,annot_kws ={'size':8},cmap = 'Blues')
```
