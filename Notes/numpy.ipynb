List Vs Numpy - Time Taken

```python
from time import process_time
```

Time taken by a list

```python
python_list =[i for i in range (10000)]

start_time = process_time()

python_list = [i+5 for i in python_list]

end_time = process_time()

print(end_time - start_time)
```

```python
np_array = np.array([i for i in range(10000)])

start_time = process_time()

np_array += 5
end_time = process_time()

print(end_time - start_time)

```



Intial Placholders in Numpy Arrays



```python
# create a numpy array of Zeros
x = np.zeros([4,5])
print(x)
```

```python

# create a numpy array of ones
y = np.ones([3,3])
print(y)
```

```python

# array of Particular Values
z = np.full((5,4),7)         # ((rows,columns),value)
print(z)
```

```python
# create an identity matrix

a = np.eye(4)
print(a)
```

```python
# create a numpy array with random Values

b = np.random.random((3,5))
print(b)
```

```python
# random integer values array with specific range

c = np.random.randint(10,100,(3,4))  # (ranges,(row ,column))
print(c)
```

```python
# array of evenly space values
d = np.linspace(10,30,6)   # (start,stop,no. of values)
print(d)
```

```python
# array of evenly spaced values --> specifiy the step

e = np.arange(10,30,6)   # (row,column,step)
print(e)
```

```python
# convert a list to a numpy array

list2 =[10,20,30,40,50]

np_array = np.asarray(list2)
print(np_array)
type(np_array)
```



Mathematical Operation on an Numpy Array

```python
list1 = [i for i in range(1,6) ]
list2 = [i for i in range(6,11)]
print(list1 + list2)        # concantenate or joins the list
```

```python
a = np.random.randint(1,10,(3,3))
b = np.random.randint(10,20,(3,3))

```

```python
print(a)
print(b)
```

```python
print(a + b)  # values of  both array added
print('-------')
print(a-b)
print('-------')
print(a*b)
print('-------')
print(a/b)
```

```python
a = np.random.randint(1,10,(3,3))
b = np.random.randint(10,20,(3,3))
```

```python
print(a)
print("-------")
print(b)
```

```python
print(np.add(a,b))  # add the values of  array
print('-------')
print( np.subtract(a,b))
print('-------')
print(np.multiply(a,b))

print('-------')
print(np.divide(a,b))
```
