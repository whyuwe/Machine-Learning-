```python
# imorting the matplotlib library
import matplotlib.pyplot as plt
```

```python
# importing numpy to get data for our plot
import numpy as np
```

```python
x = np.linspace(0,10,100) # evenly spaces 100 values of 10
y =np.sin(x)
z =np.cos(x)
```



Ploting the data

```python
# sine wave
plt.plot(x,y) # empty plot
plt.show()
```

```python
# ploting the cosine wave
plt.plot(x,z)
plt.show()
```

```python
# adding title x-axis and y-axis labels
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine wave')
plt.title('sine wave')
```

Ploting a Parbola

```python
x =np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Parbola')
plt.show()
```

```python
x =np.linspace(-10,10,20)
y = x**2
plt.plot(x,y,'r+') # r+

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Parbola')
plt.show()
```

```python
plt.plot(x,y,'g.')
```



Bar Plot

```python
fig= plt.figure()
ax = fig.add_axes([0,0,1,1])   #(area or sizw)
languages = ['english','french','Spanish','Sanskrit','German']
people =[100,50,150,25,70]
ax.bar(languages,people)
plt.xlabel('languages')
plt.ylabel('Number of People')
plt.show()
```

Pie chart

```python
fig1 =plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['english','french','Spanish','Sanskrit','German']
people =[100,50,150,25,70]
ax.pie(people,labels = languages,autopct ='%1.2f%%' )
plt.xlabel('languages')
plt.ylabel('Number of People')
plt.show()
```

Scatter Plot

```python
x =np.linspace(0,10,30)
y = np.sin(x)
z =np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color= 'g')
ax.scatter(x,z,color= 'b')
plt.xlabel('X-AXIS')
plt.ylabel('Y-AXIS')
plt.show()
```

3D Scatter plot

```python
fig4 =plt.figure()
ax = plt.axes(projection = '3d')
z = 20* np.random.random(100)
x =np.sin(z)
y =np.cos(z)
ax.scatter(x,y,z,color ='r')    #ax.scatter(x,y,z,c =z,cmap ='Blues')
plt.show()
```
