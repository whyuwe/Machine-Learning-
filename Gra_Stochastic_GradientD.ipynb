```python
# apply  Stochastic gradient descent 
class MeraSGDRegressor:
    def __init__(self,learning_rate ,epochs):
        
       self.coef =None
       self.intercept =None
       self.l_rate =learning_rate
       self.epochs =epochs
       
    def fit(self,X_train,Y_train):
        # init your coef
        self.intercept =0
        self.coef =np.ones(X_train.shape[1])
        
         
        for i in range(self.epochs):
           
            
            for j in range(X_train.shape[0]):
                idx = np.random.randint(0,X_train.shape[0])
                y_hat = np.dot(X_train[idx],self.coef) + self.intercept
                
                intercept_der =(-2)*(Y_train[idx] -y_hat)
                self.intercept = self.intercept -self.l_rate*intercept_der
                
                
                coef_der = (-2)*np.dot((Y_train[idx] -y_hat),X_train[idx])
                self.coef =self.coef -self.l_rate*coef_der
            
        return print("intercept",self.intercept),print("coef",self.coef)
            
            
            
            
            
            
        
    def predict(self,X_test):
        return np.dot(X_test,self.coef ) + self.intercept
    
    
sgd =MeraSGDRegressor(0.01,50)
        
        
        
```

```python
sgd.fit(X_train,Y_train)
```

```python
y_predict1 = sgd.predict(X_test)
```

```python
r2_score(Y_test,y_predict1)
```

# MINI BATCH GRADIENT DESCENT

```python
# apply  Mini Batch gradient descent
import random
 
class MeraMiniBGD:
    def __init__(self,learning_rate ,epochs,batch_size):
        
       self.coef =None
       self.intercept =None
       self.l_rate =learning_rate
       self.epochs =epochs
       self.batch_size =batch_size
       
    def fit(self,X_train,Y_train):
        # init your coef
        self.intercept =0
        self.coef =np.ones(X_train.shape[1])
        
         
        for i in range(self.epochs):
            
            for j in range(int(X_train.shape[0]/self.batch_size)):
                
              idx = random.sample(range(X_train.shape[0]),self.batch_size)
            # Vectorization 
              y_hat = np.dot(X_train[idx],self.coef) + self.intercept # y_predict  = mx + b
            
            
              intercept_der = (-2)*np.mean(Y_train[idx] -y_hat)
              self.intercept = self.intercept - (self.l_rate)*(intercept_der)
            
            
              coef_der = (-2)*np.dot((Y_train[idx] -y_hat),X_train[idx])/X_train[idx]  # X_train.shape[0] = the no. rows in data
            # it is one step calculation all the rows ans column show we use mertics multipilcation
            
              self.coef = self.coef -(self.l_rate)*coef_der
            
            
            
        return print("intercept",self.intercept),print("coef",self.coef)
            
            
            
            
            
            
        
    def predict(self,X_test):
        return np.dot(X_test,self.coef ) + self.intercept
    
    
Mini =MeraMiniBGD(batch_size=int(X_train.shape[0]/10),learning_rate=.01,epochs=50)
        
        
        
```
