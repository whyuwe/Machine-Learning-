# K- Means Clustering

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
df =pd.read_csv('dataset\student_clustering.csv')
```

```python
plt.scatter(df['cgpa'],df['iq'])
```

```python
from sklearn.cluster import KMeans
```

```python
wcss =[]
for i in range(1,11):
    km =KMeans(n_clusters=i)
    km.fit(df)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss) # to check elbow point
```

```python
X =df.iloc[:,:].values
km =KMeans(n_clusters=4)

km.fit(X)
```

```python
y_means = km.predict(X)
```

```python
X[y_means==3]
```

```python
X[y_means==3 ,0] # 0 indicates the first value
```

```python
plt.scatter(X[y_means== 0,0],X[y_means== 0,1],color ='blue')
plt.scatter(X[y_means== 1,0],X[y_means== 1,1],color ='red')
plt.scatter(X[y_means== 2,0],X[y_means== 2,1],color ='green')
plt.scatter(X[y_means== 3,0],X[y_means== 3,1],color ='yellow')
```

# KMeans on 3-Data

```python
from sklearn.datasets import make_blobs
centroids =[(-5,-5,5),(5,5,-5),(3.5,-2.5,4),(-2.5,2.5,-4)]

cluster_std = [1,1,1,1]

X,y = make_blobs(n_samples=200,cluster_std=cluster_std,centers=centroids,n_features=3,random_state=1)
```

```python
import plotly.express as px
fig =px.scatter_3d(x =X[:,0],y =X[:,1],z=X[:,2])
fig.show()
```

```python
km =KMeans(n_clusters=4)
km.fit(X,y)
y_means =km.predict(X)
```

```python
df1 =pd.DataFrame()
df1['col1'] =X[:,0]
df1['col2'] =X[:,1]
df1['col3'] =X[:,2]
df1['label'] =y_means
```

```python

fig = px.scatter_3d(x =df1['col1'],y =df1['col2'],z=df1['col3'],color =df1['label'])
fig.show()
```
